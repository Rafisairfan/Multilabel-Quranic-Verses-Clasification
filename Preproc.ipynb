## Penghilang Tanda Baca

```python
def remove_punctuation(text):
  text = REPLACE_BY_SPACE_RE.sub('', text) # remove symbols which are in BAD_SYMBOLS_RE from text. substitute the matched string in BAD_SYMBOLS_RE with nothing. 
  return text
```

## Case Folding

```python
def case_folding(text) :
  text = text.lower()
  
  return text
```

## Stop Word Removal

```python
stop_words = set(stopwords.words('english'))
def stopword_remove(text) :
  return " ".join([word for word in str(text).split() if word not in stop_words])
```

## Tokenizer

```python
def tokenize(text) :
  text = nltk.word_tokenize(text)
  return text
```
